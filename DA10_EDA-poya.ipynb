# plot functions:

```python
def plot_numeric_distribution(df, column):
    
    # histogram, KDE, boxplot for a numeric column.
    fig, axes = plt.subplots(2, 1, figsize=(10, 6), 
                             gridspec_kw={'height_ratios': [3, 1]})
    
    # Histogram + KDE
    sns.histplot(df[column].dropna(), bins=30, kde=True, ax=axes[0], color='skyblue')
    axes[0].set_title(f'Distribution of {column}')
    axes[0].set_xlabel(column)
    axes[0].set_ylabel('Count')
    
    # Boxplot
    sns.boxplot(x=df[column], ax=axes[1], color='lightgreen')
    axes[1].set_xlabel(column)
    
    plt.tight_layout()
    plt.show()
```

# test : mann whitney u (not normal disterbution)

```python
from scipy.stats import mannwhitneyu

approved = df[df['LoanApproved'] == 1]['TotalDebtToIncomeRatio'].dropna()
rejected = df[df['LoanApproved'] == 0]['TotalDebtToIncomeRatio'].dropna()

# Mann-Whitney U test
stat, p = mannwhitneyu(approved, rejected, alternative='two-sided')

print("p-value:", p)
```

- Since p-value < 0.05, the difference is **statistically significant**.

# “Applicants with high debt-to-income ratios are much more likely to have their loans rejected.”

# confidence interval

```python
import scipy.stats as st

def mean_confidence_interval(data, confidence=0.95):
    data = np.array(data.dropna())
    n = len(data)
    mean = np.mean(data)
    se = st.sem(data)
    h = se * st.t.ppf((1 + confidence) / 2., n-1)
    return mean, mean-h, mean+h

approved_tdir = df[df['LoanApproved'] == 1]['TotalDebtToIncomeRatio']
rejected_tdir = df[df['LoanApproved'] == 0]['TotalDebtToIncomeRatio']

approved_ci = mean_confidence_interval(approved_tdir)
rejected_ci = mean_confidence_interval(rejected_tdir)

print("Approved TotalDebtToIncomeRatio (mean, 95% CI):", approved_ci)
print("Rejected TotalDebtToIncomeRatio (mean, 95% CI):", rejected_ci)
```

- The confidence intervals do not overlap, which confirms a strong and statistically significant difference between the two groups.  
- Rejected applicants consistently have much higher TotalDebtToIncomeRatios than approved applicants, indicating that high debt relative to income is a key reason for rejection.

# t-test

```python
from scipy.stats import ttest_ind
# t-test
approved = df[df['LoanApproved'] == 1]['RiskScore'].dropna()
rejected = df[df['LoanApproved'] == 0]['RiskScore'].dropna()

stat, p = ttest_ind(approved, rejected, equal_var=False)
print("p-value:", p)

# group means
print(df.groupby('LoanApproved')['RiskScore'].mean())
```

- Since p-value < 0.05, the difference is statistically significant.
- Applicants with higher RiskScores are significantly more likely to have their loans rejected.

# “Applicants with higher RiskScores are much more likely to have their loans rejected. Both the plot and the statistical test confirm this strong relationship.”

# confidence interval

```python
import scipy.stats as st

def mean_confidence_interval(data, confidence=0.95):
    data = np.array(data.dropna())
    n = len(data)
    mean = np.mean(data)
    se = st.sem(data)  # standard error
    h = se * st.t.ppf((1 + confidence) / 2., n-1)  # margin of error
    return mean, mean-h, mean+h

approved_scores = df[df['LoanApproved'] == 1]['RiskScore']
rejected_scores = df[df['LoanApproved'] == 0]['RiskScore']

approved_ci = mean_confidence_interval(approved_scores)
rejected_ci = mean_confidence_interval(rejected_scores)

print("Approved RiskScore (mean, 95% CI):", approved_ci)
print("Rejected RiskScore (mean, 95% CI):", rejected_ci)
```

- The confidence intervals for approved and rejected groups do not overlap.  
- This confirms that the difference in RiskScores between the two groups is strong and statistically significant.  
- Rejected applicants consistently have higher RiskScores than approved applicants.

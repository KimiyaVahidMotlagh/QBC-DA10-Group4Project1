Note: These columns (InterestRate, MonthlyLoanPayment, TotalDebtToIncomeRatio, BaseInterestRate, RiskScore) are calculated after the loan decision!

Keeping them would give the model information from the future (data leakage).

we dont need them as far as i know so just remove them

```python
# remove leakage columns:
leak_cols = ['InterestRate', 'MonthlyLoanPayment', 
             'TotalDebtToIncomeRatio', 'BaseInterestRate', 'RiskScore']

df= df.drop(columns=leak_cols)
df.shape
```
